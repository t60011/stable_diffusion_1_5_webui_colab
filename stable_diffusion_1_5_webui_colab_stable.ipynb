# 使用說明

*   第一次從 Github 用 **Open in Colab** 按鈕開啟本筆記後，建議先選“**檔案**”功能表下的“**在雲端硬碟儲存副本**”，然後執行**自己雲端硬碟的副本**！

## 程式碼儲存格執行說明：

*   點擊“**程式碼儲存格(一)**”左邊的**“▶︎“播放圖示**開始執行。
*   開始執行後需要在雲端下載很多檔案，請耐心等待5~15分鐘，儲存格下方的**輸出內容**會出現 **Public WebUI Colab URL: 連線網址**。
*   結束使用前，先按“**程式碼儲存格(一)**”左邊的**“◼︎“終止圖示**停止執行，然後再執點擊“**程式碼儲存格(二)**”跟“**程式碼儲存格(三)**”的**“▶︎“播放圖示**,
就能立即備份**設定檔**跟**生成圖像**到自己的 Google 雲端硬碟。
*   **離線之前**，記得要點選右上方**連線狀態**右邊的**“▼“其他連線選項選單**下的“**中斷連線並刪除執行階段**”，否則停止連線後還是會消耗運算額度。免費額度如果耗盡，會需要等待一段時間（12-48小時），或是付費之後才能使用。
*   **意外離線或關閉視窗**但沒有刪除執行階段，只要在Colab打開**自己雲端硬碟的副本**，按右上方的**“▼“其他連線選項**選單下方“**連線至代管的執行階段**”就可以重新連線。

## 程式碼儲存格圖示說明：

*   ▶︎ ：尚未執行的**程式碼儲存格**左方顯示會顯示**播放圖示**，按一下**播放圖示**該儲存格內的程式碼才會開始執行。
*   ◼︎：執行中的**程式碼儲存格**左方顯示會顯示**終止圖示**，按一下**終止圖示**該儲存格內的程式碼就會停止執行。
*   ▼：右上方**連線**狀態右邊會出現**其他連線選項選單**，按一下**其他連線選項選單**就會出現重新線或中斷的選項。

本筆記以 Stable Diffusion 1.5 WebUI Colab 穩定版為基礎加以修改，原功能包含：

*   預先載入 [runwayml/stable-diffusion-v1-5](https://huggingface.co/runwayml/stable-diffusion-v1-5) 模型。
*   除基本擴充功能外，會預先載入 ControlNet 1.1 擴充功能以及必須的模型，因此載入時間會比較久，請耐心等候。

本腳本**新增功能**為：
*   預先安裝**中文以及雙語介面**擴充功能，但第一次套用中文介面仍需要中斷程式並重新執行。
*   備份**設定檔案**到自己的 Google雲端硬碟，以免每次要顯示中文介面都要反覆中斷執行階段。
*   備份**生成圖像**到自己的 Google雲端硬碟，以免執行階段結束後圖像被刪除。
*   可以選擇是否預先安裝CivitAI 網站上的 **RealisticVision** 跟 **ChilloutMix** 兩個額外的模型，但下載時間會更久。
